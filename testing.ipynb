```python
import numpy as np
def relu(z):
    return np.maximum(0, z)

def sigmoid(z):
    return 1 / (1 + np.exp(-z))

print(sigmoid(np.array([0,1,1,1,9])))
import matplotlib.pyplot as plt

plt.plot(np.arange(-10, 10, 0.1), sigmoid(np.arange(-10, 10, 0.1)))
plt.grid(True)
plt.show()



```
